```python
import pandas as pd
import pandas.io.data as web
import datetime
import matplotlib.pyplot as plt
import statsmodels.api as sm
matplotlib.style.use('ggplot')
%pylab inline
```

```python
start = datetime.datetime(1900,1,1)
end = datetime.datetime(2016,1,1) #look up now()
a = web.DataReader("NCBEILQ027S","fred",start,end)
b = web.DataReader("NCBTCMDODNS","fred",start,end)
c = web.DataReader("TCMILBSHNO","fred",start,end)
d = web.DataReader("FGTCMDODNS","fred",start,end)
e = web.DataReader("SLGTCMDODNS","fred",start,end)
f = web.DataReader("FBCELLQ027S","fred",start,end)
g = web.DataReader("WCMITCMFODNS","fred",start,end)
spx = web.get_data_yahoo("^GSPC",start,end, interval='m')
```

```python
#Calculate 10Yr Forward Return
spx['Quarterly'] = spx['Adj Close'].pct_change(periods = 119)
#Align Data
spx['10yrForward'] = spx['Quarterly'].shift(-119)


#Change Dates to 1 to match FRED Data
spx.index = spx.index.map(lambda t: t.replace(day=1))
spx.head()
```

```python
#Transformation
#x = ((a+f)/1000)/(((a+f)/1000)+b+c+d+e+g)
x = pd.DataFrame()
x['Sum'] = ((a['NCBEILQ027S']+f['FBCELLQ027S']) / 1000)
x['Model'] = x['Sum'] / (x['Sum']+b['NCBTCMDODNS']+c['TCMILBSHNO']+d['FGTCMDODNS']+e['SLGTCMDODNS']+g['WCMITCMFODNS'])
x['Model'].plot()
```
